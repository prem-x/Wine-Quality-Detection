# üç∑ Objective 2: Perform Exploratory Data Analysis (EDA) to Uncover Patterns

```python
# Explore data characteristics
eda_summary = {
    'Missing Values': df.isnull().sum(),
    'Unique Values': df.nunique(),
    'Mean': df.mean(),
    'Median': df.median(),
    'Standard Deviation': df.std()
}
# Convert the summary into a DataFrame for better visualization
eda_df = pd.DataFrame(eda_summary)
print(" EDA Summary:")
print(eda_df)

```

# üç∑ Objective 3: Analyze Feature Relationships Using Correlation and Covariance

```python
import seaborn as sns
import matplotlib.pyplot as plt



# Calculate correlation matrix
correlation_matrix = df.corr()

# Create correlation heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5, linecolor='gray')


plt.title(" Correlation Heatmap Between Features", fontsize=16)
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()


```

```python
# Also display covariance matrix
print("üìè Covariance Matrix:")
print(df.cov())
```

# üç∑ Objective 4: Detect and Interpret Outliers in the Dataset

```python


# Step 1: Select numeric columns only
numeric_df = df.select_dtypes(include=['number'])

# Step 2: Use IQR method to detect outliers
Q1 = numeric_df.quantile(0.25)
Q3 = numeric_df.quantile(0.75)
IQR = Q3 - Q1

# Step 3: Count outliers for each numeric feature
outliers = ((numeric_df < (Q1 - 1.5 * IQR)) | (numeric_df > (Q3 + 1.5 * IQR))).sum()


# Step 4: Visualize with boxplots
plt.figure(figsize=(14, 6))
sns.boxplot(data=numeric_df, palette="Set2", linewidth=1.5)
plt.title("Boxplot for Outlier Detection", fontsize=16)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Step 4: Print number of outliers
print(" Number of Outliers Detected (IQR Method):")
print(outliers)
```

# üç∑ Objective 5: Visualize the Data Using Graphs and Charts for Insights

```python
# Distribution of wine quality
plt.figure(figsize=(10, 6))
sns.histplot(df['quality'], bins=10, kde=True, color="skyblue")
plt.title(" Distribution of Wine Quality", fontsize=16)
plt.xlabel("Quality")
plt.ylabel("Frequency")
plt.grid(True)
plt.tight_layout()
plt.show()


```

```python
# Pairplot of selected features
sns.pairplot(df[['alcohol', 'sulphates', 'citric acid', 'volatile acidity', 'quality']], hue="quality", palette="husl")
plt.suptitle(" Pairplot of Selected Features", y=1.02, fontsize=16)
plt.show()
```

```python
# Step 1: Ensure 'pH' and 'alcohol' columns are numeric
df['pH'] = pd.to_numeric(df['pH'], errors='coerce')
df['alcohol'] = pd.to_numeric(df['alcohol'], errors='coerce')

# Step 2: Create pH buckets
df['pH_bucket'] = pd.cut(
    df['pH'],
    bins=[2.9, 3.1, 3.3, 3.5, 3.7, 3.9],
    labels=["2.9-3.1", "3.1-3.3", "3.3-3.5", "3.5-3.7", "3.7-3.9"]
)

# Step 3: Pivot table for heatmap with observed=False to suppress warning
pivot = df.pivot_table(
    values='alcohol',
    index='quality',
    columns='pH_bucket',
    aggfunc='mean',
    observed=False  
)

# Step 4: Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(pivot, annot=True, cmap='YlOrBr', linewidths=0.5, linecolor='gray')
plt.title(" Average Alcohol by Wine Quality & pH Buckets", fontsize=14)
plt.xlabel("pH Range")
plt.ylabel("Wine Quality")
plt.tight_layout()
plt.show()
```
